`check dau bk 2021`

```python
def mark2percent(i,data,dfpercent):
    for loca in range(data.shape[0]):
        mark=round(data.get(i)[loca],2)
        try:
            m2p(i,data,mark,dfpercent,loca)        
        except: 
            m2pf(i,data,mark,dfpercent,loca)
def m2p(i,data,mark,dfpercent,loca):
    j=f'percent {i}'
    percent=dfpercent.get(mark)[j]
    data.loc[loca,i]=round(percent,4)
def m2pf(i,data,mark,dfpercent,loca):
    if mark>dfpercent.columns.min():
        mark=round(mark-0.05,2)
        try:m2p(i,data,mark,dfpercent,loca)
        except: m2pf(i,data,mark,dfpercent,loca)
    else: 
        data.loc[loca,i]=0
    
for i in lst:
    mark2percent(i,w_df_2021,trainT)
    
```
